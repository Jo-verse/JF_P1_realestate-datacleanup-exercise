#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
plt.figure(figsize=(10, 6))

for population in southern_belt['level5'].unique():
    population_data = southern_belt[southern_belt['level5'] == population]
    population_prices = population_data[['price']]
    population_prices_normalized = scaler.fit_transform(population_prices)
    plt.hist(population_prices_normalized, bins=30, alpha=0.5, label=population)

plt.legend(loc='upper right')
plt.title('Histograms of Normalized Prices by Town')
plt.xlabel('Normalized Price')
plt.ylabel('Frequency')
plt.show()

# TODO: The histograms show that some towns, such as Getafe, have lower and more concentrated prices, 
# while others like Fuenlabrada have a greater price spread, indicating more diversity in the available properties.
```

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph. You must use the subset obtained in Exercise 14 (★★☆)

```python
## HERE: plot the coordinates of the estates

## PUT HERE YOUR CODE:
import pandas as pd
import folium
from folium.plugins import MarkerCluster

# Load data
ds = pd.read_csv('assets/real_estate.csv', sep=';')

# Filter data
southern_belt = ds[ds['level5'].isin(['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón'])]

# Convert coordinates
southern_belt = southern_belt.copy()
southern_belt['latitude'] = pd.to_numeric(southern_belt['latitude'], errors='coerce')
southern_belt['longitude'] = pd.to_numeric(southern_belt['longitude'], errors='coerce')

# Remove rows with missing coordinates
southern_belt = southern_belt.dropna(subset=['latitude', 'longitude'])

# Define colors for each city
colors = {
    'Fuenlabrada': 'red',
    'Leganés': 'blue',
    'Getafe': 'green',
    'Alcorcón': 'purple'
}

# Create a map centered on the southern belt area
map = folium.Map(location=[40.28, -3.82], zoom_start=12)

# Create a MarkerCluster layer
marker_cluster = MarkerCluster().add_to(map)

# Add markers to the map
for _, row in southern_belt.iterrows():
    color = colors.get(row['level5'], 'black')  # Default to black if city not in colors dict
    folium.Marker(
        location=(row['latitude'], row['longitude']),
        popup=row['level5'],
        icon=folium.Icon(color=color, icon='info-sign')
    ).add_to(marker_cluster)

# Display the map
map
```
